```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('/content/drive/MyDrive/low-contast.png')
image_1 = cv2.imread('/content/drive/MyDrive/high-contrast.png')
```

```python
def plot_cv_img(input_image, output_image):
    fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(30, 40))
    ax[0].imshow(cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB))
    ax[0].set_title('Input Image', fontsize=16)
    ax[0].axis('off')
    ax[1].imshow(cv2.cvtColor(output_image, cv2.COLOR_BGR2RGB))
    ax[1].set_title(f'Decreased contrast', fontsize=16)
    ax[1].axis('off')
    plt.show()
```

```python
def decrease_contrast(image, a, s, t):
  new_image = np.zeros_like(image)
  new_image = image - s 
  new_image[new_image < 0] = 0
  new_image = (a * new_image + t).astype(np.uint8)
  new_image[new_image > 255] = 255
  return new_image
```

```python
new_image = decrease_contrast(image, 0.75, 255, 3)
plot_cv_img(image, new_image)
```

```python
new_image_1 = decrease_contrast(image, 0.9, 190, 1)
plot_cv_img(image, new_image_1)
```

```python
new_image_2 = decrease_contrast(image_1, 0.75, 1, 2)
plot_cv_img(image_1, new_image_2)
```

```python
new_image_3 = decrease_contrast(image_1, 0.75, 250, 3)
plot_cv_img(image_1, new_image_3)
```
